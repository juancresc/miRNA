```python
import pandas as pd
```

```python
file_de_srna = 'data/files/diffexpr-results.0.05.csv'
df_de_srna = pd.read_csv(file_de_srna,sep=",")
print('Total DE sRNA: ',len(df_de_srna.index))
```

```python
file_srna = 'data/21dpi_tgac/ShortStack_D.gff3'
df_srna = pd.read_csv(file_srna,sep="\t")
df_srna.columns = ['seqname', 'source', 'feature', 'start', 'end', 'score', 'strand', 'frame', 'attribute']
print('Total sRNA: ',len(df_srna.index))
```

```python
gene_list = df_de_srna.Gene.tolist()
def filter(s):
    s = s.split(';')[0].split('=')[1]
    if s in gene_list:
        return True
    return False

#shortstack gff results that are DE
df_de_srna_ann = df_srna[df_srna.attribute.apply(lambda s: filter(s))]
print('Total sRNA DE filtered: ',len(df_de_srna_ann.index))
```

```python
file_genes = 'data/cross/genes.gff'
df_genes = pd.read_csv(file_genes,sep="\t",header=None)
df_genes.columns = ['seqname', 'source', 'feature', 'start', 'end', 'score', 'strand', 'frame', 'attribute']
df_genes['gene'] = df_genes['attribute'].str.split(';').apply(lambda x: x[0]).str.split(':').apply(lambda x: x[1])
print('Total genes: ',len(df_genes.index))
```

```python
df_de_srna_ann['cluster'] = df_de_srna_ann.attribute.str.split(';').apply(lambda x: x[0]).str.split('=').apply(lambda x: x[1])
```

```python
#DE genes that are close to DE sRNA
df_de_srna_ann['gene'] = ''
distance = 1500
for k,s in df_de_srna_ann.iterrows():
    start = s.start - distance
    end = s.end + distance
    df_res = df_genes[(df_genes.seqname == s.seqname) & (df_genes.end >= start) & (df_genes.start <= end)]
    if len(df_res.index) == 0:
        continue
    gene = df_res.gene.tolist()[0]
    df_de_srna_ann.loc[k, 'gene'] = gene
```

```python
df_de_srna_ann
```

```python
df_de_srna_ann_merge = df_de_srna_ann
file_mart_ae = 'data/mart/aestivum.txt'
df_mart = pd.read_csv(file_mart_ae, delimiter='\t', quotechar = '"',escapechar='\\')
df_mart.rename(columns={'Gene stable ID': 'gene'}, inplace=True)
df_de_srna_ann_merge = pd.merge(df_de_srna_ann_merge, df_mart, on='gene',how = 'left')

file_mart_ae = 'data/mart/orthologues.txt'
df_mart = pd.read_csv(file_mart_ae, delimiter=',', quotechar = '"',escapechar='\\')
df_mart.rename(columns={'Gene stable ID': 'gene'}, inplace=True)
df_de_srna_ann_merge = pd.merge(df_de_srna_ann_merge, df_mart, on='gene',how = 'left')

file_mart_ae = 'data/mart/tauschii.txt'
df_mart = pd.read_csv(file_mart_ae, delimiter='\t', quotechar = '"',escapechar='\\')
df_mart.rename(columns={'Gene description': 'Tauschii'}, inplace=True)
df_mart.rename(columns={'Gene stable ID': 'Aegilops tauschii gene stable ID'}, inplace=True)
df_de_srna_ann_merge = pd.merge(df_de_srna_ann_merge, df_mart, on='Aegilops tauschii gene stable ID',how = 'left')

file_mart_ae = 'data/mart/thaliana.txt'
df_mart = pd.read_csv(file_mart_ae, delimiter=',', quotechar = '"',escapechar='\\')
df_mart.rename(columns={'Gene description': 'Thaliana'}, inplace=True)
df_mart.rename(columns={'Gene stable ID': 'Arabidopsis thaliana gene stable ID'}, inplace=True)
df_de_srna_ann_merge = pd.merge(df_de_srna_ann_merge, df_mart, on='Arabidopsis thaliana gene stable ID',how = 'left')

```

```python
df_de_srna_ann_merge.to_csv('data/files/sRNA_DE_genes.csv',sep='\t',index=None)
```
